**과제1**

```python
movies = pd.read_csv("pandas_data/movie.csv")
movie1= movies.iloc[:, :11].fillna(0)
```

```python
'''
실수형 변환
num_critic_for_reviews = float32
duration = float32
director_facebook_likes = float32
actor_3_facebook_likes = float32
actor_1_facebook_likes = float32
gross = float32

num_critic_for_reviews, duration, director_facebook_likes, actor_3_facebook_likes의 경우 
float16으로 변환 가능하지만 기술통계로 인해 float32로 변환
'''
movie1.describe().T
```

```python
movie1 = movie1.astype(
    {"num_critic_for_reviews":np.float32,
     "duration":np.float32,
     "director_facebook_likes":np.float,
     "actor_3_facebook_likes":np.float32,
     "actor_1_facebook_likes":np.float32,
     "gross":np.float32}
)
```

```python
movie1.info()
```

```python
movie1.memory_usage(deep = True)
```

```python
# color열만 category형으로 변환
movie1.describe(include = ["object"]).T
```

```python
# color의 경우 3가지의 unique값이 있는걸로 특정 구간이 반복된다. 그렇기에 category형으로 변환 가능
movie1 = movie1.astype(
    {"color":"category"}
)
```

```python
movie1.memory_usage(deep = True)
```

**과제2**

```python
college = pd.read_csv("pandas_data/college.csv")
```

```python
college[["CITY", "UGDS"]].sort_values(['UGDS'], ascending = False).drop_duplicates(subset = 'CITY')
```

```python
# 위 식과 끝 부분이 다른 이유는 아마도 NaN값을 기준으로 도시 이름이 정렬된 듯하다.

college[["CITY", "UGDS"]].groupby('CITY', as_index = False).apply(lambda df: df.sort_values(
    'UGDS', ascending = False
    ).head(1)
).droplevel(0).sort_values('UGDS', ascending = False)                                                              
```

**과제3**

```python
# 주별 UGDS의 최대값

college[["STABBR", "UGDS"]].groupby('STABBR', as_index = False).apply(lambda df: df.sort_values(
    'UGDS', ascending = False
    ).head(1)
).droplevel(0).sort_values('UGDS', ascending = False)  
```

```python
# RELAFFIL별 UGDS의 최대값

college[["RELAFFIL", "UGDS"]].groupby('RELAFFIL', as_index = False).apply(lambda df: df.sort_values(
    'UGDS', ascending = False
    ).head(1)
).droplevel(0).sort_values('UGDS', ascending = False)  
```
